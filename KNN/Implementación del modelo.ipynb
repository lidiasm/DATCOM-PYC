```r
## Seleccion aleatoria de columnas

rm(list=ls())
dataTrain1 = read.csv("dataTrainMinMaxEtiqueta1")
dataTest = read.csv("dataTestMinMax")

y1 = as.factor(dataTrain1$h1n1_vaccine)
y1
dataTrain1$h1n1_vaccine = as.factor(y1)
str(dataTrain1)
dataTrain1 = select(dataTrain1,-c(seasonal_vaccine))
head(dataTrain1)

mat_val_h1n1 = c()
v = c('h1n1_concern', 'h1n1_knowledge', 'behavioral_antiviral_meds',
     'behavioral_avoidance', 'behavioral_face_mask', 'behavioral_wash_hands',
     'behavioral_large_gatherings', 'behavioral_outside_home',
     'behavioral_touch_face', 'doctor_recc_h1n1', 'doctor_recc_seasonal',
     'chronic_med_condition', 'child_under_6_months', 'health_worker',
     'opinion_h1n1_vacc_effective', 'opinion_h1n1_risk',
     'opinion_h1n1_sick_from_vacc', 'opinion_seas_vacc_effective',
     'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'age_group',
     'education', 'sex', 'income_poverty', 'marital_status', 'rent_or_own',
     'employment_status', 'census_msa', 'household_adults',
     'household_children',  'race_White',
     'race_Black', 'race_Other.or.Multiple', 'race_Hispanic',"h1n1_vaccine")

  



  mat_val_h1n1 = c()
  for(i in 1:100){
    print(i)
    val_ale = sample(34,16,replace=FALSE)
    model_h1n1 = kknn(h1n1_vaccine~.,
                      train = dataTrain1[v[c(val_ale,35)]],
                      test = dataTest[v[val_ale]], k = 51, distance = 1,  scale=FALSE)
    
    h1n1_predict = model_h1n1$prob[,2]
    mat_val_h1n1 = cbind(mat_val_h1n1,as.numeric(as.character(h1n1_predict)))
  }
  
  
  v = c('h1n1_concern', 'h1n1_knowledge', 'behavioral_antiviral_meds',
        'behavioral_avoidance', 'behavioral_face_mask', 'behavioral_wash_hands',
        'behavioral_large_gatherings', 'behavioral_outside_home',
        'behavioral_touch_face', 'doctor_recc_h1n1', 'doctor_recc_seasonal',
        'chronic_med_condition', 'child_under_6_months', 'health_worker',
        'opinion_h1n1_vacc_effective', 'opinion_h1n1_risk',
        'opinion_h1n1_sick_from_vacc', 'opinion_seas_vacc_effective',
        'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'age_group',
        'education', 'sex', 'income_poverty', 'marital_status', 'rent_or_own',
        'employment_status', 'census_msa', 'household_adults',
        'household_children',  'race_White',
        'race_Black', 'race_Other.or.Multiple', 'race_Hispanic',"seasonal_vaccine")
  
  dataTrain2 = read.csv("dataTrainMinMaxEtiqueta2")
  y2 = as.factor(dataTrain2$seasonal_vaccine)
  y2
  dataTrain2$seasonal_vaccine = as.factor(y2)
  str(dataTrain2)
  dataTrain2 = select(dataTrain2,-c(h1n1_vaccine))
  head(dataTrain2)
  mat_val_seasonal= c()
  
  for(i in 1:100){
    print(i)
    val_ale = sample(32,16,replace=FALSE)
    v[c(val_ale,35)] %in% names(dataTrain2)
    model_seasonal <- kknn(seasonal_vaccine~.,
                           train = dataTrain2[v[c(val_ale,35)]],
                           test = dataTest[v[val_ale]], k = 51, distance = 1,  scale=FALSE)
    seasonal_predict = model_seasonal$prob[,2]
    mat_val_seasonal = cbind(mat_val_seasonal,as.numeric(as.character(seasonal_predict)))
  }

  
  test_id = c(26707:53414)
  val_seasonal = apply(mat_val_seasonal,1,mean)
  val_h1n1 = apply(mat_val_h1n1,1,mean)
  pred10 = data.frame(test_id,val_h1n1,val_seasonal)
  names(pred10) = c("respondent_id","h1n1_vaccine","seasonal_vaccine")
  write.csv(pred10,"ensemble100R.csv",row.names=FALSE)

```
