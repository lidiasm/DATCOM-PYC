```r
#Resumen estadístico y estudio del número de variables y clase de las mísmas
summary(dataf)
str(dataf)
```

```r
#Estudio inicial de los valores perdidos 
aggr_plot <- aggr(dataf, col=c('blue','red'), numbers=TRUE,sortVars=TRUE, labels=names(dataf), cex.axis=.7, gap=3,
                   ylab=c("Histogram of missing data","Pattern"))
```

```r
#unión del conjunto train con las etiquetas de las dos clases a predecir
joined_df = merge(dataf,labels)
joined_df= joined_df[,-c(1,16,35,36)]
ncol(joined_df)
str(joined_df)
```

```r
#volvemos a estudiar los missing values con las etiquetas
aggr_plot <- aggr(joined_df, col=c('blue','red'), numbers=TRUE,sortVars=TRUE, labels=names(dataf), cex.axis=.7, gap=3,
                   ylab=c("Histogram of missing data","Pattern"))
```

```r
nrow(joined_df)
```

```r
#imputación de valores perdidos con mice
imputed_Data = mice(joined_df, m=5, maxit = 5, method = 'pmm', seed = 500)
```

```r
df = complete(imputed_Data)
```

```r
#comprobamos que no quede ningún valor perdido
aggr_plot <- aggr(df, col=c('blue','red'), numbers=TRUE,sortVars=TRUE, labels=names(df), cex.axis=.7, gap=3,
                   ylab=c("Histogram of missing data","Pattern"))
```

```r
write.csv(df,"C:/Users/javij/Desktop/test_clean.csv", row.names = FALSE)
```
